# Preparando Dataset de Turmas com dados dos Alunos da ECT

```python
import pandas as pd
```

```python
df_turmas = pd.DataFrame()
```

```python
# Concatenando todas as turmas de 2016
f = open('../turmas_new/turmas-20161.csv')
d = f.read()
d = d.replace("\n", "")
x = d.split("\";\"")
a = []
step = 25
for b in range(0, len(x), step):
    a.append(x[b:b + step])
df = pd.DataFrame(a[1:], columns=['id_turma'] + a[0][1:])
df_turmas = pd.concat([df_turmas, df])
```

```python
# conversoes
df_turmas['id_turma'] = df_turmas.id_turma.astype(int)
df_turmas['id_componente_curricular'] = df_turmas.id_componente_curricular.astype(int)
df_turmas['ano'] = df_turmas.ano.astype(int)
df_turmas['periodo'] = df_turmas.periodo.astype(int)
```

```python
df_turmas.shape
```

# Relacionando com Componentes Curriculares

```python
# Lendo tabelas de com nomes dos componentes
df_componentes = pd.read_csv('../componentes-curriculares-presenciais-ok.csv')
```

```python
# Pegando apenas da ECT
#df_componentes = df_componentes[df_componentes.unidade_responsavel == "ESCOLA DE CIÊNCIAS E TECNOLOGIA"]
#df_componentes = df_componentes[df_componentes.curso_componente == "CIÊNCIAS E TECNOLOGIA"]

# Pegando apenas colunas interessantes
df_componentes = df_componentes[['id_componente', 'nome']]
```

```python
# Merge entre Datasets
df_turmasNome = pd.merge(df_turmas, df_componentes, left_on='id_componente_curricular', right_on='id_componente')
```

```python
df_turmasNome.shape
```

```python
# selecionando colunas de Turmas
colunas = ['id_turma', 'nome', 'ano', 'periodo']
df_turmasNome = df_turmasNome[colunas]
```

```python
df_turmasNome.head()
```

```python
# Baixando Dataset de Matriculas
df_matriculas = pd.read_csv("../data/matriculas_2015-2017.csv")
```

```python
df_auxAll = pd.merge(df_matriculas, df_turmasNome, on='id_turma')
df_auxAll = df_auxAll.drop_duplicates().dropna()
```

```python
materias = ['CÁLCULO I', 'VETORES E GEOMETRIA ANALÍTICA', 'PRÉ-CÁLCULO',
               'QUÍMICA GERAL', "PRÁTICAS DE LEITURA E ESCRITA", "CIÊNCIA, TECNOLOGIA E SOCIEDADE"]
```

```python
df_auxTurmas = df_auxAll[df_auxAll["nome"].isin(materias)]
```

```python
df_auxTurmas = df_auxTurmas.iloc[:, 1:]
df_auxTurmas
```

```python
df_auxTurmas.to_csv("../data/dados_ect.csv", index=False)
```
